# Problem 5 : Like/Unlike posts
Like/Unlike posts

1.Liking the top 30 posts of the ‘dilsefoodie'. Print message if you have already liked it.

2.Unliking the top 30 posts of the ‘dilsefoodie’. Print message if you have already unliked it.

```python
#Question 5 --> search the profile 'dilsefoodie' and liked and unlike its top 30 posts

time.sleep(2)
driver.find_element(By.CSS_SELECTOR,'[aria-label=Search]').click() #opening the search bar by clicking
search_input = driver.find_element(By.CLASS_NAME,'x1lugfcp')       #loacting the search bar
search_input.send_keys('dilsefoodie')                              #sending the profile name 'dilsefoodie'

time.sleep(3)
driver.find_element(By.XPATH,'//div[contains(@class,"x6s0dn4 x78zum5 xdt5ytf x5yr21d x1odjw0f x1n2onr6 xh8yej3")]/div/a').click()
                                                                   #locating the 'dilsefoodie' profile and clicking on it
```

```python
# Question 5 continues
time.sleep(3)                                                      #waiting for the page to load
driver.execute_script('window.scrollTo(0,3000);')                  #scrolling down manually
time.sleep(3)                                                      
driver.execute_script('window.scrollTo(0,3000);')                  #scrolling down manually till 30 posts are not loaded
time.sleep(3)                                                      #waiting to load
posts = driver.find_elements(By.CLASS_NAME,'_aagw')                #extracting the posts into posts list 

# for liking the 30 posts 
for i in range(30):                                                #the process for liking starts
    posts[i].click()                                               #opening each posts by clicking it  
    time.sleep(2)                                                  #waiting for the post to load 
    Like = driver.find_element(By.CLASS_NAME,'_aamw')              #loacting the like button  
    st = BeautifulSoup(Like.get_attribute('innerHTML'),'html.parser').title.string # taking its text like or Unlike
    if st == 'Like':                                               #if the text is Like then it means that the post is not liked   
        Like.click()                                               #likng the post by clicking the button
        time.sleep(2)
    else:
        print("Already liked post number",i+1)                     #is the post is already liked then printing 'alredy liked' with the post number
        time.sleep(2)
    driver.find_element(By.CSS_SELECTOR,'[aria-label=Close]').click() #closing the post by clicking the close button in top right corner
    if i == 29:
        print('All 30 liked')                                      #as the there are not exactly 30 posts in our list so if posts reach the limit of 30 then we break
        break                                                      #and printing that liked all the 30 posts
```

```python
#Question 5 continues
time.sleep(3)                                                      #same process goes for the unliking process the change is the everywhere instead for like there unlike
driver.execute_script('window.scrollTo(0,3000);')
time.sleep(3)
driver.execute_script('window.scrollTo(0,3000);')
time.sleep(3)
posts = driver.find_elements(By.CLASS_NAME,'_aagw')
for i in range(30):
    posts[i].click()
    time.sleep(2)
    unlike = driver.find_element(By.CLASS_NAME,'_aamw')
    st = BeautifulSoup(unlike.get_attribute('innerHTML'),'html.parser').title.string
    if st == 'Unlike':
        unlike.click()
        time.sleep(2)
    else:
        print("Already Unliked it",i+1)
        time.sleep(2)
    driver.find_element(By.CSS_SELECTOR,'[aria-label=Close]').click()
    if i == 29:
        print('All 30 unliked')
        break
```
